**Task 1 — Credit Scoring Business Understanding (Summary)**
- Basel II impact: Requires interpretable, auditable models and strong documentation for capital and supervisory review.
- Proxy target: We use RFM-based proxy labels because true default labels are unavailable; this enables model development but introduces proxy risk and potential bias — monitor, validate, and retrain when real defaults arrive.
- Model trade-offs: Use Logistic Regression with WoE for production explainability and regulatory compliance; use Gradient Boosting as a validation/benchmark model to detect additional signal and interactions.
- Action items: Monitor proxy alignment, implement conservative thresholds and appeals, collect real default outcomes for retraining, and maintain MLflow audit trails.
